```python
# get high measurements
user_freq = load_data['connect_time'].value_counts(ascending=True).nlargest(10).index
```

```python
load_data['downthpt'][load_data['connect_time'] == user_freq[0]].plot()
```

```python
def getThroughput():
    throughputValues = []
    for index, value in enumerate(user_freq):
        throughput = load_data['downthpt'][load_data['connect_time'] == user_freq[index]]
        throughputValues.append(list(pd.Series(throughput, index=None)))
        
        #user_data = pd.concat(throughputValues, axis=0, ignore_index=True)
        #user_data = user_data.reset_index(drop=True)
        
    return throughputValues
```

```python
throughput_of_clients = getThroughput()
```

```python
clients = ['client 01', 'client 02', 'client 03', 'client 04', 'client 05',
           'client 06', 'client 07', 'client 08', 'client 09','client 10']
df = pd.Series(throughput_of_clients, index=[clients])
```

```python
# plt.figure(figsize=(10, 8))
sns.set(rc={"figure.figsize":(10, 4)})
fig = sns.boxplot(data=df, color=".8", linewidth=.75);

#plt.title("Customized Boxplots")
plt.xlabel("Clients")
plt.ylabel("Throughput [Mbps]")

plt.show();
```
